## Regression 線性迴歸

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
```

```python
x_data = np.linspace(0, 1, 101) # x_data 0~1 shape [101]
noise = np.random.normal(scale=0.1, size=x_data.shape) # 雜訊!! 
k = 0.87
b = 0.17

y_data = k*x_data + b + noise # y_data shape [101]
```

```python
plt.plot(x_data,y_data,'.')
plt.ylim(0, 1.2)
```

```python
b = tf.Variable(tf.zeros(1))
k = tf.Variable(tf.zeros(1))
y = k * x_data + b

#平方差和 
loss = tf.reduce_mean((y_data - y)**2)
#梯度下降 learn Rate 0.5
optimizer = tf.train.GradientDescentOptimizer(0.08)
#optimizer = tf.train.MomentumOptimizer(0.02, 0.9)
#optimizer = tf.train.AdagradOptimizer(0.1)
#最小化loss
train = optimizer.minimize(loss)
#變數初始化
init_op = tf.global_variables_initializer()


# sess = tf.Session()
# writer = tf.summary.FileWriter("logs/", sess.graph)
```

```python
with tf.Session() as sess:
    sess.run(init_op)
    klst = []
    blst = []
    for step in range(1001):
        sess.run(train)
        klst.append(sess.run(k))
        blst.append(sess.run(b))
        if step%100 == 0:
            kk, bb = sess.run([k, b])
            print('{:4d} k:{:4.2f} b:{:4.2f}'.format(step, kk.item(), bb.item()))
```

```python
line_x = np.linspace(0, 1, 3)
line_y = line_x*kk + bb
plt.plot(x_data, y_data,'.')
plt.plot(line_x, line_y,'-r')
plt.ylim(0, 1.2)
```

```python
#創建維度要高於一資料一階 方便廣播運算 : shape : (b值, k值, 計算值)
line_k = np.linspace(0, 1, 101).reshape(-1, 1)
line_b = np.linspace(0, 1, 101).reshape(-1, 1 ,1)

arr_y = ((line_k * x_data) + line_b)
loss_arr = ((y_data - arr_y)**2).mean(2)
```

### k, b 的loss 等高線圖

```python
plt.contour(loss_arr, 50, extent=[0, 1, 0, 1], cmap=plt.cm.coolwarm)
plt.colorbar()
plt.xlabel('k')
plt.ylabel('b')
plt.plot(klst, blst,'.g')
```

### L2 正規化(範例)

### k, b 的loss 等高線圖  loss加上L2正規化 變得有曲線

```python
arr_y = ((line_k * x_data) + line_b) + (s * (line_k**2 + line_b**2) / 2)
loss_arr = ((y_data - arr_y)**2).mean(2)
plt.contour(loss_arr, 50, extent=[0, 1, 0, 1], cmap=plt.cm.coolwarm)
plt.colorbar()
plt.plot(klst, blst,'.g')
```

### L2正規化只是範例 本案例不適合使用正規化, 正規化可以有效的讓參數不過大

```python
x_data = np.linspace(-3, 3 ,100, dtype=np.float32)
noise = np.random.normal(0, 0.2, x_data.shape)
xk = 0.6
xb = -0.7

y_data = xk*x_data**2 + xb + noise
```

```python
plt.plot(x_data, y_data, '.')
```

```python
b = tf.Variable(2.5)
k = tf.Variable(-2.4)
y = tf.multiply(k ,tf.square(x_data)) + b

#平方差和 
loss = tf.reduce_mean(tf.square(y_data - y))
#梯度下降 learn Rate 0.5
optimizer = tf.train.GradientDescentOptimizer(0.005)
#最小化loss
train = optimizer.minimize(loss)
#變數初始化
init_op = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init_op)
    klst = []
    blst = []
    for step in range(1001):
        sess.run(train)
        klst.append(sess.run(k))
        blst.append(sess.run(b))
        if step%100 == 0:
            kk, bb = sess.run([k, b])
            print('{:4d} k:{:4.2f} b:{:4.2f}'.format(step, kk.item(), bb.item()))
```

```python
yline = kk*x_data**2 + bb
plt.plot(x_data, y_data, '.')
plt.plot(x_data, yline, '-r')
```

```python
x_data = np.linspace(0, 1 , 5, dtype=np.float32)
noise = np.random.normal(0, 0.2, x_data.shape)
xk = 0.4
xb = 0.2

y_data = xk*x_data + xb + noise

plt.plot(x_data, y_data, '.')
```

```python
aa, bb, cc, dd = 1., 1., 1., 1.
a = tf.Variable(aa)
b = tf.Variable(bb)
c = tf.Variable(cc)
d = tf.Variable(dd)
y = a + b*x_data + c*x_data**2 + d*x_data**3

#平方差和 
loss = tf.reduce_mean(tf.square(y_data - y))
#梯度下降 learn Rate 0.5
optimizer = tf.train.GradientDescentOptimizer(0.6)
#最小化loss
train = optimizer.minimize(loss)
#變數初始化
init_op = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init_op)
    for step in range(10001):
        sess.run(train)
        if step%2000 == 0:
            print(step, sess.run([a, b, c, d]))
    aa, bb, cc, dd =sess.run([a, b, c, d])
```

```python
xline = np.linspace(0, 1, 101)
yline = aa + bb*xline + cc*xline**2 + dd*xline**3
plt.plot(x_data, y_data, '.')
plt.plot(xline, yline, '-r')
```
